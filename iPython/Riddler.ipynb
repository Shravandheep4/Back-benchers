```python
subjectname = input("Subject name: ")
raw_corpus = u""

textfile = sorted(glob.glob("Resources\\" + subjectname + "\\" + "\*.txt"))

for text in textfile:
    print(f"Reading '{text}'")
    
    with codecs.open(text, "r", "utf-8") as file:
        raw_corpus += file.read()
     
clean = re.sub("[^a-zA-Z\.]", " ", raw_corpus)
raw_corpus = clean.split()
raw_corpus = " ".join(raw_corpus)
print(raw_corpus)
```

```python
text = raw_corpus or 'Hinton is a British cognitive psychologist and computer scientist, most noted for his work on artificial neural networks. Hinton was one of the first researchers who demonstrated the use of generalized backpropagation algorithm for training multilayer neural nets. He is a leading figure in the deep learning community. Hinton is called by some as the "Godfather of Deep Learning."'

tokenizer = nltk.data.load('tokenizers/punkt/english.pickle')
sentences = tokenizer.tokenize(text)
```

```python
cachedStopWords =  nltk.corpus.stopwords.words("english")

def sentence_to_wordlist(raw):
    clean = re.sub("[^a-zA-Z\.]", " ", raw)
    words = clean.split()
    stemmer = nltk.PorterStemmer()
    
    filtered_sentence = [w for w in words if not w in cachedStopWords]
    return " ".join(filtered_sentence)
```

```python
filtered_sentences = []

for sentence in sentences:
    sentence = sentence_to_wordlist(sentence)
    text = nltk.word_tokenize(sentence)
    text = nltk.pos_tag(text)
    filtered_sentences.append(text)
    #print(text, end="\n\n") 
```

```python
#Define your grammar using regular expressions
grammar = ('''
    NP: {<DT>?<JJ>*<NN>} # NP
    ''')
chunkParser = nltk.RegexpParser(grammar)

for text in filtered_sentences:    
    tree = chunkParser.parse(text)
    tree.draw()
    break
```

# After performing speech tagging and chuncking, We frame the questions
